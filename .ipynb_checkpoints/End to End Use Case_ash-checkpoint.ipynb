```python
# Lets do Univariate Histograms 
def histograms(data):
    data.hist()
    plt.tight_layout()
    plt.show()
```

```python
histograms(data)
```

```python
# Lets do side by side histplot adn boxplots 
def histplot_boxplot(data, feature, figsize=(12,7), bins=None):
    fig, (ax_box, ax_hist)=plt.subplots(
    nrows=2,
    sharex=True,
    gridspec_kw={"height_ratios":(0.25, 0.75)},
    figsize=figsize )
    
    sns.boxplot(data=data, x=feature, showmeans=True, color="violet", ax=ax_box)
    sns.histplot(data=data, x=feature,pallete="winter", bins=bins, ax=ax_hist) if bins else sns.histplot(data=data,
                                                                     x=feature, ax=ax_hist)
    ax_hist.axvline(data[feature].mean(), linestyle="--", color="green")
    ax_hist.axvline(data[feature].median(), linestyle="-", color="black")
    plt.show()
```

```python
for col in data.select_dtypes(exclude="O").columns:
    print(col)
    histplot_boxplot(data=data, feature=col)
    plt.show()
    
    
```

```python
# univariate Countplot
def countplot(data, feature):
    plt.figure(figsize=(12,7))
    ax=sns.countplot(data=data, x=feature, color="green")
    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate("{:.3g}%".format(100.*y/len(data)), (x.mean(), y), ha="center", va="bottom")
    plt.show()
```

```python
countplot(data=data,feature="Outcome")
```

```python
# Bivariate -categorical vs numerical
def bivariate_barplot(data, feature1, feature2):
    data.groupby(feature1)[feature2].mean().plot(kind="bar", color="orange")
    plt.ylabel(col)
    plt.show()
```

```python
for col in data.select_dtypes(exclude="O").columns:
    print("Outcome vs ", col)
    bivariate_barplot(data=data, feature1="Outcome",feature2= col)
    print("**************************************")
```

Those with higher Pregnancies, Glucose, Insulin, BMI and Age are more likely to be diabetic

```python
# numericall vs numerical
def lineplot_scatterplot(data, feature1, feature2):
    plt.figure(figsize=(16,7))
    plt.subplot(1,2,1)
    sns.lineplot(data=data, x=feature1, y=feature2, color="green")
    plt.title("Lineplotbetween {0} and {1}".format(feature1, feature2))
    
    plt.subplot(1,2,2)
    sns.scatterplot(data=data, x=feature1, y=feature2, color="orange", hue="Outcome")
    plt.title("Scatter Plot Between {0} and {1}".format(feature1, feature2))
    plt.show()
```

```python
for col in data.select_dtypes(exclude="O").columns:
    lineplot_scatterplot(data=data, feature1="Age", feature2=col)
    
    
```

# To be contd....

```python
df=data.copy()
df.head()
```

```python
x=df.drop('Outcome',1)
y=df['Outcome']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
xtrain
```

```python
ytrain.replace({'No':0,'Yes':1},inplace=True)
```

```python
ytest.replace({'No':0,'Yes':1},inplace=True)
```

```python
sc=StandardScaler()

xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
xtrain.shape,xtest.shape
```

```python
# create a metrics function

def print_metric(ytest,ypred,model_name):
    print('Model Name ',model_name)
    print('')
    print('Accuracy score {:.3f}'.format(accuracy_score(ytest,ypred)))
    print('')
    print('Recall score {:.3f}'.format(recall_score(ytest,ypred)))
    print('')
    print('Precission score {:.3f}'.format(precision_score(ytest,ypred)))
    print('')
    print('F1 score {:.3f}'.format(f1_score(ytest,ypred)))
    print('')
    print('ROC-AUC score {:.3f}'.format(roc_auc_score(ytest,ypred)))
    print('')
    print('Classification Report')
    print(classification_report(ytest,ypred))
```

```python
def plot_metrics(clf, xtest,ytest):
    skm.plot_confusion_matrix(clf,xtest,ytest,display_labels=[0,1])
    print('')
    skm.plot_roc_curve(clf,xtest,ytest)
    print('')
    skm.plot_precision_recall_curve(clf,xtest,ytest)
    plt.show()
    
```

```python
# fit and evaluate a knn model

knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
print_metric(ytest,ypred,'KNN')
```

```python
plot_metrics(knn,xtest,ytest)
```

```python
#Draw a model complexity plot to finalize k
neighbours=range(1,25,1)
train=np.empty(len(neighbours))
test=np.empty(len(neighbours))
for i,k in enumerate(neighbours):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtrain,ytrain)
    ypred=knn.predict(xtest)
    train[i]=knn.score(xtrain,ytrain)
    test[i]=knn.score(xtest,ytest)
    
plt.title('KNN Model Complexity Plots')
plt.plot(neighbours,train,label='Train Accuracy')
plt.plot(neighbours,test,label='Test Accuracy')
plt.legend()
plt.xlabel('Neighbours')
plt.ylabel('Accuracy')
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=13)
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
print_metric(ytest,ypred,'KNN')
```

```python
plot_metrics(knn,xtest,ytest)
```

```python
#Draw a model complexity plot to finalize k
neighbours=range(1,25,1)
train=np.empty(len(neighbours))
test=np.empty(len(neighbours))
for i,k in enumerate(neighbours):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtrain,ytrain)
    ypred=knn.predict(xtest)
    train[i]=knn.score(xtrain,ytrain)
    test[i]=knn.score(xtest,ytest)
    
plt.title('KNN Model Complexity Plots')
plt.plot(neighbours,train,label='Train Accuracy')
plt.plot(neighbours,test,label='Test Accuracy')
plt.legend()
plt.xlabel('Neighbours')
plt.ylabel('Accuracy')
plt.show()
```

```python
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
print_metric(ytest,ypred,'Logreg')
```

```python
#optimize
c=np.logspace(-5,13,180)
param_grid={'C':c}
n_fold=10
cv=GridSearchCV(estimator=logreg,param_grid=param_grid,scoring='f1',verbose=1,n_jobs=-1,cv=n_fold,return_train_score=False)
cv.fit(xtrain,ytrain)
cv.best_score_
```

```python
logreg_cv=cv.best_estimator_
logreg_cv.fit(xtrain,ytrain)
ypred=logreg_cv.predict(xtest)
print_metric(ytest,ypred,'Logreg_cv')
```
